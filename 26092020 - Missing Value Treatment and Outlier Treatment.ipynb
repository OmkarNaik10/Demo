## Imputation of Continuous Values

```python
data.absences.head() # first 5 rows for absences col
```

```python
data.absences.hist(bins = 20);
```

#### Since, the data is skewed, we will use the median to impute the missing values

```python
data.absences.median()
```

- The fillna() function is used to fill NA/NaN values using the specified method.

```python
data.absences = data.absences.fillna(data.absences.median())
data.absences.head()
```

## Imputation for Categorical Variables

```python
data.activities.head(10)
```

```python
data.activities.mode()
```

```python
data.activities.mode()[0]
```

```python
data.activities = data.activities.fillna(data.activities.mode()[0])
data.activities.head(10)
```

```python
data.activities.isnull().sum()
```
