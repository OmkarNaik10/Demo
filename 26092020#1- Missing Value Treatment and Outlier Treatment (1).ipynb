# Outliers

```python
data.boxplot(column = 'absences');
```

**(Q1 - 1.5 * IQR) > = data < = (Q3 + 1.5 * IQR)**

```python
min_value = data['absences'].min()
Q1 = data['absences'].quantile(0.25) # 25% data is less than Q1
median_value = data['absences'].median()
Q3 = data['absences'].quantile(0.75) # 75% data is less than Q3
max_value = data['absences'].max()

print ("min_value    :", min_value)
print ("Q1           :", Q1)
print ("median_value :", median_value)
print ("Q3           :", Q3)
print ("max_value    :", max_value)
```

```python
IQR = Q3 - Q1
print (IQR)
```

```python
lower_limit = Q1 - 1.5 * IQR # 1.5 is used in this formula. Normal distribution considering 99% of data
upper_limit = Q3 + 1.5 * IQR

# data points less than lower_limit are outliers
# data points greater than upper limit are outliers

print (lower_limit)
print ()
print (upper_limit)
```

### Deleting Outliers

```python
data[data.absences <= lower_limit]
```

```python
data[data.absences >= upper_limit]
```

### Imputing the Outlier

```python
def imputer(value):
    if value < lower_limit or value > upper_limit:
        return median_value
    else:
        return value

data.absences = data.absences.apply(imputer)
```

```python
data.boxplot(column = 'absences');
```

```python
# After the outlier treatment

min_value = data['absences'].min()
Q1 = data['absences'].quantile(0.25)
Q3 = data['absences'].quantile(0.75)
max_value = data['absences'].max()
median_value = data['absences'].median()

print ("min_value    :", min_value)
print ("Q1           :", Q1)
print ("median_value :", median_value)
print ("Q3           :", Q3)
print ("max_value    :", max_value)
```

lines that join min to Q1 and max value to Q3 are whiskers
